```python
from pathlib import Path
from PIL import Image
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from numpy import asarray
# image_lr = Image.open(image_file).resize((W_pad_lr, H_pad_lr), Image.BILINEAR)
images_root="/home/phdcs2/Hard_Disk/Datasets/Deep-Spectral-Segmentation/data/object-segmentation/ECSSD/images"
image_id = '0125_resized'
image_file = str(Path(images_root) / f'{image_id}.png')
```

```python
dir='/home/phdcs2/Hard_Disk/Datasets/Deep-Spectral-Segmentation/data/object-segmentation/ECSSD/ground_truth_mask/'
image_file = str(Path(dir) / f'{image_id}.png')
image = Image.open(image_file, mode='r')
img = asarray(image)
img_rgb = cv.cvtColor(img,cv.COLOR_GRAY2RGB)
print(image.format)
print(image.size)
print(image.mode)
image.show()
```

```python
import matplotlib.pyplot as plt
# numpy_array = W_feat.numpy()

# Downsample the tensor by averaging values in small blocks
downsampled_array = np.mean(img.reshape(128, 4, 128, 4), axis=(1, 3))
# downsampled_array = np.mean(W_feat.reshape(64, 16, 64, 16), axis=(1, 3))

# Plot the heatmap of the downsampled tensor
plt.imshow(downsampled_array, cmap='YlGnBu', interpolation='nearest')
plt.colorbar()
plt.title('Original Patch Feature')
plt.show()
```

```python
# img=cv.imread(image_file)
print(type(img))
print(img.shape)
kernel = np.ones((15,15),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv.bilateralFilter(img,15,75,75)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Bilateral Filtering')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

median = cv.medianBlur(img,15)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv.GaussianBlur(img,(15,15),0)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Gaussian Blurring')
plt.xticks([]), plt.yticks([])
plt.show()
```
